# Cleaning of DATA

There are two methods that the data set can be cleaned. 
1. The first method is to remove the rows with missing values. 
2. The second method is to fill the missing values with the mean/median of the column. 

But first, the data types must first be converted to float because the data types are currently objects

Create **Array** for month columns

```python
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

Convert **Monthly_Columns** data type into numerical data types and
make sure to coerce missing/mismatched values into **NaN**

```python
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df[monthly_columns] = df[monthly_columns].apply(pd.to_numeric, errors='coerce')
```

Check Data types and **NaN** values

```python
df.info()
```

```python
print(df[monthly_columns].isnull().sum())
```

Now That data type objects for the monthly columns have been converted into numerical data types, it is now time to clean the data set with one of the two methods that have been mentioned recently.

The second method seems to be more appropriate for this specific kind of scenario because the missing/mismatched data can be inputted with the Mean/Median of the certain column in order to utilize all of the data while having *minimal effect* as opposed to completely erasing that specific row rendering **All** of the data to be **unused**.

## Replacing **NaN** values with *Median*

The reason there are **NaN** values may mean that the equipment that day performed to be faulty and was unable to get an accurate translation of the data into the *API/AQI* for that specific month/period

We replace values with the Median instead of the Mean to account for extreme values for each City.

Create a copy of the data set

```python
df_x=df

df_x.info()
```

```python
df_x.head()
```

in order to replace the **NaN** with the Median for that certain city/country, the **Lambda function** will be utilized.

Order of replacing:
1. City
2. Country
3. Global Median - this is due to some countries being distinct and acts as a general fallback should there be remaining invalid values. This would only apply to very little spaces so the difference is negligible.

```python
df_x[monthly_columns] = df_x.groupby("City")[monthly_columns].transform(lambda x: x.fillna(x.mean()))
df_x[monthly_columns] = df_x.groupby("Country")[monthly_columns].transform(lambda x: x.fillna(x.median()))
df_x[monthly_columns] = df_x[monthly_columns].fillna(df[monthly_columns].median())
df_x[monthly_columns] = df[monthly_columns].round(2)
df_x.info()
```

Check for **NaN** values

```python
print(df_x[monthly_columns].isnull().sum())
```
