# Distributions, Trends, Patterns and Visualizations

With the Given Data, We can now utilize graphs to visualize trends, patters and outliers within the data set

```python
df.describe()
```

```python
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

Create a **historgraph** to display values for each month

```python
sns.set_style("whitegrid")

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 20))  
axes = axes.flatten()  

for i, col in enumerate(monthly_columns):
    if i < len(axes): 
        sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color="royalblue")
        axes[i].set_title(f"Distribution of {col}", fontsize=12)
        axes[i].set_xlabel("")
        axes[i].set_ylabel("API/AQI", fontsize=10)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

For the average AQI/API of each month, a line plot will be utilized

```python
df[monthly_columns].mean().plot(kind='line', 
                                marker='o',
                                figsize=(15, 8), 
                                color = 'royalblue',)
plt.title('Monthly Average API/AQI', fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.ylabel('API/AQI', fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.xlabel('Month', fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.xticks(ticks=range(len(monthly_columns)), labels=monthly_columns)
plt.grid(axis='both', linestyle='-', alpha=1)
plt.show()
```

Create a **Boxplot** to compare the variety for every month as well as find any extremities such as outliers.

```python
plt.figure(figsize=(12, 12))
sns.boxplot(data=df[monthly_columns])
plt.title("Monthly Air Quality Variability", fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("API/AQI", fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.xlabel("Month", fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.show()
```

Get the 10 most and least polluted cities

```python
top_polluted = df[['City', 'Country', 'Year']].sort_values(by='Year', ascending=False).head(10)
print(top_polluted)
```

```python
least_polluted = df[['City', 'Country', 'Year']].sort_values(by='Year', ascending=True).head(10)
print(least_polluted)
```

```python
plt.figure(figsize=(13, 8))
plt.bar(top_polluted['City'], 
        top_polluted['Year'], 
        color=['royalblue','cyan','lightgreen','lightcoral','orange','purple','pink','yellow','red','blue'])

plt.grid(axis='x')
plt.xlabel("City",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("Air Quality Index (Yearly)",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.title("Top 10 Most Polluted Cities in Asia for the Year 2023",fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.ylim(0, top_polluted['Year'].max() + 10)  


for i, value in enumerate(top_polluted['Year']):
    plt.text(i, value + 1, f"{value:.1f}", ha='center', fontsize=12, fontweight='bold',fontfamily='tahoma')

plt.show()
```

```python
plt.figure(figsize=(13, 8))
plt.bar(least_polluted['City'], 
        least_polluted['Year'], 
        color=['royalblue','cyan','lightgreen','lightcoral','orange','purple','pink','yellow','red','blue'])

plt.grid(axis='x')
plt.xlabel("City",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("Air Quality Index (Yearly)",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.title("Top 10 Least Polluted Cities in Asia for the Year 2023",fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.ylim(0, least_polluted['Year'].max() + 10)  


for i, value in enumerate(least_polluted['Year']):
    plt.text(i, value + 1, f"{value:.1f}", ha='center', fontsize=12, fontweight='bold',fontfamily='tahoma')

plt.show()
```

Get the top 10 most and least polluted **countries**

```python
top_polluted_country = df[['Country', 'Year']].groupby('Country').mean().sort_values(by='Year', ascending=False).head(10).round(2).reset_index()
print(top_polluted_country)
```

```python
least_polluted_country = df[['Country', 'Year']].groupby('Country').mean().sort_values(by='Year', ascending=True).head(10).round(2).reset_index()
print(least_polluted_country)
```

```python
plt.figure(figsize=(15, 8))
plt.bar(top_polluted_country['Country'], 
        top_polluted_country['Year'], 
        color=['royalblue','cyan','lightgreen','lightcoral','orange','purple','pink','yellow','red','blue'])

plt.grid(axis='x')
plt.xlabel("Country",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("Air Quality Index (Yearly)",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.title("Top 10 most Polluted Countries in Asia for the Year 2023",fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.ylim(0, top_polluted_country['Year'].max() + 10)  


for i, value in enumerate(top_polluted_country['Year']):
    plt.text(i, value + 1, f"{value:.1f}", ha='center', fontsize=12, fontweight='bold',fontfamily='tahoma')

plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.bar(least_polluted_country['Country'], 
        least_polluted_country['Year'], 
        color=['royalblue','cyan','lightgreen','lightcoral','orange','purple','pink','yellow','red','blue'])

plt.grid(axis='x')
plt.xlabel("Country",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("Air Quality Index (Yearly)",fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.title("Top 10 least Polluted Countries in Asia for the Year 2023",fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.ylim(0, least_polluted_country['Year'].max() + 10)  


for i, value in enumerate(least_polluted_country['Year']):
    plt.text(i, value + 1, f"{value:.1f}", ha='center', fontsize=12, fontweight='bold',fontfamily='tahoma')

plt.show()
```

Create a Heatmap analysis to see if Seasons matter for the AQI/API

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[monthly_columns].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation of Monthly Air Quality", fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.show()
```

Conclusion 
 
Air quality tends to remain stable across adjacent months but shifts notably in mid-year, likely due to seasonal changes. <br>
Winter months (NovFeb) show strong correlation, suggesting persistent pollution effects. <br>
Mid-year (AprilJuly) breaks the pattern, possibly due to rain or weather changes. <br>
Seasons do affect the monthly AQI/API for every country/city.

Create a boxplot to look for extreme outliers for every city

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Country", y="Year", data=df)
plt.xticks(rotation=90)
plt.title("Country-wise Air Quality Distribution", fontsize=15 , fontfamily='tahoma', fontweight='bold')
plt.xlabel("Country", fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.ylabel("Air Quality Index (Yearly)", fontsize=12 , fontfamily='tahoma', fontweight='bold')
plt.show()
```

Based on the data found, the air quality in Asia is generally poor. The air quality index (AQI) is high in most cities and countries in Asia.
The average Air quality is highest during January and December, which could be explained that seasons have a drastic effect when it comes to pollution. Countries mostly in India show the highest average as well as extreme outliers within their respective cities. Meanwhile Japan, China and Singapore show to be the least polluted countries in Asia on the year 2023.
