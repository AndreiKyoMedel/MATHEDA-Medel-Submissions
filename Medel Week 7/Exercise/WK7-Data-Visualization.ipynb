Import **Pandas**, **numpy**, and **matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

## Number of Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind='bar',
                        color = ['lightblue' , 'blue' , 'darkblue' , 'skyblue' , 'cyan'],
                        figsize=(15,15))

chart.bar_label(chart.containers[0], 
                label_type= 'edge',
                fontfamily = 'Tahoma',
                fontsize = 15,
                fontweight = 'bold')

plt.title('Chips Manufactured (CPU + GPU)',
          fontfamily = 'Tahoma',
          fontsize = 15,
          fontweight = 'bold')

plt.xlabel('Type of manufacturer',
           fontfamily = 'Tahoma',
           fontsize = 15,
           fontweight = 'bold')

chart.set_xticklabels(chip_count.index, rotation=0,
            fontfamily = 'Tahoma',
            fontsize = 10)

plt.ylabel('Number of Chips',
            fontfamily = 'Tahoma',
            fontsize = 15,
            fontweight = 'bold')
plt.show()
```

## Numbers of CPUs Manufactured

```python
df_cpu = df_w[df_w['Type'] == 'CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python
chart = cpu_count.plot(kind='bar',figsize=(10,10), color = ['black' , 'yellow'])

plt.title('CPUs Manufactured',
          fontfamily = 'Tahoma',
          fontsize = 10,
          fontweight = 'bold')

plt.ylabel('CPU Count',
            fontfamily = 'Tahoma',
            fontsize = 10,
            fontweight = 'bold')

plt.xlabel('Type of Manufacturer',
            fontfamily = 'Tahoma',
            fontsize = 15,
            fontweight = 'bold')

plt.show()
```

## Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by = 'TDP')
df_tdp
```

```python
df_tdp = df_tdp.head(7)
df_tdp
```

```python
plt.figure(figsize=(10,5))

plt.bar(df_tdp['Product'], df_tdp['TDP'])

plt.title('Products with the Lowest Thermal Display Power', 
          fontfamily = 'Tahoma', 
          fontsize = 10, 
          fontweight = 'bold')
plt.ylabel('Wattage')

plt.xlabel('Product Name/Model', 
           fontfamily = 'Tahoma', 
           fontweight = 'bold')

plt.xticks(fontsize = 7 , ha='center')
plt.show()
```

## GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_ps['Year'] , df_ps['Process_Size'])
plt.title('GPU Process Size Trend',
            fontfamily = 'Tahoma',
            fontsize = 15,
            fontweight = 'bold')
plt.xlabel('Year Processed',
            fontfamily = 'Tahoma',
            fontsize = 15,
            fontweight = 'bold')
plt.ylabel('Process Size')
```
